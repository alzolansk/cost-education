```python
# 📦 Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import requests
from datetime import date
import numpy as np
import locale
```

```python
# 📦 Requests
df = pd.read_csv('../data/International_Education_Costs.csv', usecols=['Country', 'City', 'University', 'Program', 'Level', 'Duration_Years', 'Tuition_USD', 'Rent_USD', 'Insurance_USD', 'Insurance_USD'])
url = 'https://economia.awesomeapi.com.br/json/last/USD-BRL'
resposta = requests.get(url).json()
cotacao_USD = float(resposta['USDBRL']['bid'])
cotacao_USD = np.round(cotacao_USD, 2)
locale.setlocale(locale.LC_ALL, 'Portuguese_Brazil.1252')
#---------------------------------------------------
df['Tuition_BRL'] = (df['Tuition_USD'] * cotacao_USD).round().astype(int)
df['Rent_BRL'] = (df['Rent_USD'] * cotacao_USD).round().astype(int)
df['Insurance_BRL'] = (df['Insurance_USD'] * cotacao_USD).round().astype(int)
```

```python
df['Custo_Total_BRL'] = (
    df['Tuition_BRL'] +
    (df['Rent_BRL'] * df['Duration_Years'] * 12) +
    (df['Insurance_BRL'] * df['Duration_Years'])
).astype(int)
```

```python
#🧑‍🎓 Média de gastos dos estudantes por país
media_paises = df.groupby('Country')['Custo_Total_BRL'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=media_paises.values, y=media_paises.index, hue=media_paises.index, palette='viridis')

for i, valor in enumerate(media_paises.values):
    texto_formatado = locale.currency(valor, grouping=True)
    plt.text(
        x=valor + 100,
        y=i,
        s=texto_formatado,
        va='center'
    )
    
plt.title('Top 10 países com maior custo total em BRL')
plt.xlabel('Custo médio total (R$)')
plt.ylabel('País')
plt.xlim(0, media_paises.max() + 150_000)
plt.show()
```

```python
# Estados Unidos tem a maior média de gastos entre os países analisados, e representa mais que o dobro da média de Israel (R$326 mil)
# Apesar do custo alto, os EUA oferecem mais bolsas para estudantes internacionais do que países como Hong Kong ou Singapura.
```

```python
media_brasil = df.query('"Brazil" in Country')['Custo_Total_BRL'].mean().round(2)
media_paises = df.groupby('Country')['Custo_Total_BRL'].mean().sort_values(ascending=False)
media_paises_top10 = media_paises.head(10)

comparacao = media_paises_top10.copy()
comparacao['Brazil'] = media_brasil
comparacao = comparacao.sort_values(ascending=False)
ranking_brasil = media_paises.rank(ascending=False, method='min').loc['Brazil']
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=comparacao.index, y=comparacao.values, palette='magma', hue=comparacao.index)

# Rótulos
for i, valor in enumerate(comparacao.values):
    plt.text(
        x=i,
        y=valor + 10000,
        s=locale.currency(int(valor), grouping=True),
        ha='center',
        fontsize=8
    )

    plt.title(f'Comparação do Brasil com os países de maior custo total\n (O Brasil é um dos mais baratos, na posição {int(ranking_brasil)} de {len(media_paises)} países.)')
plt.xlabel('País')
plt.ylabel('Custo médio total (R$)')
plt.ylim(0, comparacao.max() + 150_000)
plt.xticks(rotation=45)
plt.tight_layout()

plt.axhline(
    y=media_brasil,       # posição vertical da linha
    color='red',          # cor vermelha
    linestyle='--',       # linha tracejada
    linewidth=2           # espessura da linha
)

plt.text(
    x=len(comparacao) - 1,                     # coloca o texto perto da direita
    y=media_brasil + 50000,                    # um pouco acima da linha vermelha
    s=f'Média Brasil', 
    color='red',
    ha='right'
)

plt.show()
```

```python
# A média de gastos para estudantes no Brasil é de R$ 98.145, enquanto nos Estados Unidos ultrapassa R$ 722 mil — ou seja, mais de 7x maior.
```
